```python
def convert_to_pianoroll_2(work):
    k = work.analyze('key')
    i = m21.interval.Interval(k.tonic, m21.pitch.Pitch('C'))
    wrk = work.flat.transpose(i)
    
    dic = {}
    for i in range(int(wrk.highestTime*4)):
        dic[i]=[]
    for n in wrk.notes:
        off = int(n.offset*4)
        d = n.duration
        dur = int(d.quarterLength*4)
        for pit in n.pitches:
            mpit = pit.midi
            for i in range(off,off+dur):
                dic[i].append(mpit)
    return dic
```

```python
d = convert_to_pianoroll_2(bwv295)
```

```python
s1 = m21.stream.Stream()
```

```python
for notes in d:
    for mpit in d[notes]:
        n = m21.note.Note()
        n.pitch.midi = mpit
        n.quarterLength = 1
        offset = notes//4
        s1.insert(offset, n)
        print(f'{notes} {n.offset} {n.pitch.midi}')
```
